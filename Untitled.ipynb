## Fix a few labels in the given data set 

```python
df = df.rename(columns={' Temperature':'Avg_Temperature',
                        ' Average humidity (%': 'Avg_humidity',
                        ' Average dewpoint (°F':'Avg_dewpoint',
                        ' Average barometer (in':'Avg_barometer',
                        ' Average windspeed (mph':'Avg_windspeed',
                        ' Average gustspeed (mph':'Avg_gustspeed',
                       ' Average direction (°deg':'Avg_direction',
                       ' Rainfall for month (in' : 'Rainfall_in_a_month',
                       'Rainfall for year (in' : 'Rainfall_in_a_year',
                        ' Maximum rain per minute' : 'Max_rain_per_min',
                       ' Maximum temperature (°F' : 'Max_Temp', 
                       ' Minimum temperature (°F' : 'Min_Temp',
                       ' Maximum humidity (%' : 'Max_humidity',
                       ' Minimum humidity (%' : 'Min_humidity',
                        ' Maximum pressure' : 'Max_pressure',
                        ' Minimum pressure' : 'Min_pressure',
                       ' Maximum windspeed (mph' : 'Max_windspeed',
                       ' Maximum gust speed (mph' : 'Max_gust_speed',
                       ' Maximum heat index (°F' : 'Max_heat_index'})
```

```python
df=df.drop(['Date1'],axis=1)
df=df.drop([' Month'],axis=1)
```

```python
df["Date"] = df["Date"].str.replace("'","")
df["Avg_Temperature"]=df["Avg_Temperature"].str.replace("'","")
df["Avg_humidity"] = df["Avg_humidity"].str.replace("'","")
df["Avg_dewpoint"] = df["Avg_dewpoint"].str.replace("'","")
df["Avg_barometer"] = df["Avg_barometer"].str.replace("'","")
df["Avg_windspeed"] = df["Avg_windspeed"].str.replace("'","")
df["Avg_gustspeed"] = df["Avg_gustspeed"].str.replace("'","")
df["Avg_direction"] = df["Avg_direction"].str.replace("'","")
df["Rainfall_in_a_month"] = df["Rainfall_in_a_month"].str.replace("'","")
df["Rainfall_in_a_year"] = df["Rainfall_in_a_year"].str.replace("'","")
df["Max_rain_per_min"] = df["Max_rain_per_min"].str.replace("'","")
df["Max_Temp"] = df["Max_Temp"].str.replace("'","")
df["Min_Temp"] = df["Min_Temp"].str.replace("'","")
df["Max_humidity"] = df["Max_humidity"].str.replace("'","")
df["Min_humidity"] = df["Min_humidity"].str.replace("'","")
df["Max_pressure"] = df["Max_pressure"].str.replace("'","")
df["Min_pressure"] = df["Min_pressure"].str.replace("'","")
df["Max_windspeed"] = df["Max_windspeed"].str.replace("'","")
df["Max_gust_speed"] = df["Max_gust_speed"].str.replace("'","")
df["Max_heat_index"] = df["Max_heat_index"].str.replace("'","")
df[" diff_pressure"] = df[" diff_pressure"].str.replace("'","")
```

```python
df.columns
```

```python
# Removal of irrelevant data
df = df[df['Date'] != '2022-02-29']
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df = df.reset_index(drop=True)
```

## Year Correction

```python
target_date = '2022-01-01'
target_date = pd.to_datetime(target_date)
index = df.index[df['Date'] == target_date].tolist()
year = 2021
max_year = 2033

for i in range(index[0], df.shape[0]):
    if df.at[i,'Date'] == target_date:
        year += 1
        if year > max_year:
            break
    df.at[i,'Date'] = df.at[i,'Date'].replace(year=year)
```
